```python
import pandas as pd
dataFrame = pd.read_csv('interstellar_travel.csv')
dataFrame
```

```python
print('Ukuran Data: ', dataFrame.shape)
```

```python
feature_names = dataFrame.columns[:-1]
target_names = 'Loyalty Program Member'

features = dataFrame[feature_names]
target = dataFrame[target_names]

print('Feature Names :', features)
print('Target Names :', target)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

dataFrame['Gender']   = label_encoder.fit_transform(dataFrame['Gender'])
dataFrame['Occupation']   = label_encoder.fit_transform(dataFrame['Occupation'])
dataFrame['Travel Class']   = label_encoder.fit_transform(dataFrame['Travel Class'])
dataFrame['Destination']   = label_encoder.fit_transform(dataFrame['Destination'])
dataFrame['Star System']   = label_encoder.fit_transform(dataFrame['Star System'])
dataFrame['Purpose of Travel']   = label_encoder.fit_transform(dataFrame['Purpose of Travel'])
dataFrame['Transportation Type']   = label_encoder.fit_transform(dataFrame['Transportation Type'])
dataFrame['Booking Date']   = label_encoder.fit_transform(dataFrame['Booking Date'])
dataFrame['Departure Date']   = label_encoder.fit_transform(dataFrame['Departure Date'])
dataFrame['Special Requests']   = label_encoder.fit_transform(dataFrame['Special Requests'])
dataFrame['Loyalty Program Member']   = label_encoder.fit_transform(dataFrame['Loyalty Program Member'])

dataFrame
```

```python
X = dataFrame.drop('Loyalty Program Member', axis=1)
Y = dataFrame['Loyalty Program Member']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
from sklearn import metrics
print('Accuracy Score of K-NN Model: ', metrics.accuracy_score(y_test, y_predict))
```
