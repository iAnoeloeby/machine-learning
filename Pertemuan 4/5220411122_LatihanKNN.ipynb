```python
import pandas as pd
dataFrame = pd.read_csv('BankChurn.csv')
dataFrame
```

```python
dataSample = dataFrame.sample(10000)
dataSample
```

```python
print('Ukuran Data: ', dataSample.shape)
```

```python
# dataSample.columns
```

```python
feature_names = dataSample.columns[:-1]
target_names = 'Exited'

features = dataSample[feature_names]
target = dataSample[target_names]

print('Feature Names :', features)
print('Target Names :', target)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

dataSample['Surname']   = label_encoder.fit_transform(dataSample['Surname'])
dataSample['Geography'] = label_encoder.fit_transform(dataSample['Geography'])
dataSample['Gender']    = label_encoder.fit_transform(dataSample['Gender'])

dataSample
```

```python
X = dataSample.drop('Exited', axis=1)
Y = dataSample['Exited']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
from sklearn import metrics
print('Accuracy of K-NN Model: ', metrics.accuracy_score(y_test, y_predict))
```
