```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv('laptop_data_cleaned.csv')
dataFrame
```

```python
dataFrame.dropna()
```

**Data Cleaning Service**

```python
dataFrame.dropna()
dataFrame.drop_duplicates(inplace=True)

dataFrame
```

```python
dataFrame['Os'].value_counts()
```

**Label Encoder**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

dataFrame['Company']    = label_encoder.fit_transform(dataFrame['Company'])
dataFrame['TypeName']   = label_encoder.fit_transform(dataFrame['TypeName'])
dataFrame['Ram']        = label_encoder.fit_transform(dataFrame['Ram'])
dataFrame['Weight']     = label_encoder.fit_transform(dataFrame['Weight'])
dataFrame['Price']      = label_encoder.fit_transform(dataFrame['Price'])
dataFrame['Ppi']        = label_encoder.fit_transform(dataFrame['Ppi'])
dataFrame['Cpu_brand']  = label_encoder.fit_transform(dataFrame['Cpu_brand'])
dataFrame['HDD']        = label_encoder.fit_transform(dataFrame['HDD'])
dataFrame['SSD']        = label_encoder.fit_transform(dataFrame['SSD'])
dataFrame['Gpu_brand']  = label_encoder.fit_transform(dataFrame['Gpu_brand'])
dataFrame['Os']         = label_encoder.fit_transform(dataFrame['Os'])

dataFrame
```

```python
dataFrame['Os'].value_counts()
```

**Balancing Data With SMOTE**

```python
from imblearn.over_sampling import SMOTE

target_features     = dataFrame.drop("Os", axis=1)
label_features      = dataFrame['Os']

#SMOTE: balance the dataset
smote = SMOTE(random_state=42)
X_resampled, Y_resampled = smote.fit_resample(target_features, label_features)

print("Class distribution after SMOTE:")
# print(pd.Series(balance_dataFrame).value_counts())
Y_resampled.value_counts()
```

```python
X = X_resampled
Y = Y_resampled
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

**Decision Tree Classifier Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score  

clf = DecisionTreeClassifier(random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Skor Akurasi Decision Tree :", accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
import numpy as np

newData = np.array([[20, 2, 3, 4, 2, 4, 0, 4, 9, 12, 3, 12]])

scaledData = scaler.fit_transform(newData)
preds = clf.predict(scaledData)
print("Skor Decision Tree Class:", preds)
```
